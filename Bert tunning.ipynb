```python
import pandas as pd
from transformers import BertTokenizer
import tensorflow as tf
```

```python
dataset = pd.read_csv('kaggle_ecommerce_complete_Rank_Crowd_AGG_Descriptions_ALL_ROWS (1) - kaggle_ecommerce_complete_Rank_Crowd_AGG_Descriptions_ALL_ROWS (1).csv')
```

```python
dataset.head()
```

```python
dataset.drop(['Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'product_link', 'product_price', 'url', 'product_image', 'source'], axis=1, inplace=True)
```

```python
dataset.groupby('query').count()
```

```python
print('Loading BERT tokenizer...')
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
```

```python
dataset = dataset.dropna(subset=['query', 'relevance', 'relevance:variance', 'product_title'])
```

```python
dataset
```

```python
dataset['query title'] = dataset['query'] + ' [SEP] ' + dataset['product_title']
```

```python
max_len = 0
input_ids = []
attention_masks = []

def tokenize_sentence(sentence):
    global input_ids, attention_masks 
    encoded_dict = tokenizer.encode_plus(
                        sentence,                      # Sentence to encode.
                        add_special_tokens = True, # Add '[CLS]' and '[SEP]'
                        max_length = 256,       # Pad & truncate all sentences.
                        pad_to_max_length = True,
                        return_attention_mask = True,   # Construct attn. masks.
                        return_tensors = 'pt',     # Return pytorch tensors.
                   )
    input_ids.append(encoded_dict['input_ids'])
    attention_masks.append(encoded_dict['attention_mask'])

dataset['query title'].map(lambda sentence: tokenize_sentence(sentence))
```
